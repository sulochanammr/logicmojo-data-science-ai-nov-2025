```python
import numpy as np
import matplotlib.pyplot as plt
class TriangularDistribution:
    def __init__(self, minimum, maximum, mode):
        self.minimum = minimum
        self.maximum = maximum
        self.mode = mode
    def pdf(self, x):
        a = self.minimum
        b = self.maximum
        c = self.mode
        y_val = []
        if type(x) is list:
            for x_val in x:
                if a <= x_val and x_val < c:
                    y_val.append (2.0 * (x_val - a) /((b - a) * (c - a)))
                elif c < x_val and x_val <= b:
                    y_val.append (2.0 * (b - x_val) / ((b - a) * (b - c)))
                else:
                     y_val.append(0.0)
            return y_val
        else:
            if a <= x and x < c:
               return 2.0 * (x - a) /((b - a) * (c - a))
            elif c < x and c <= b:
                return 2.0 * (b - x) / ((b - a) * (b - c))
            else:
                return 0.0
    def plot_pdf(self):
        x = np.linspace(self.minimum, self.maximum, 10)
        x=x.tolist()
        y=TriangularDistribution.pdf(x)
        plt.plot(x, y, label='Graph', color='blue', linestyle='--') 
        plt.title('Graph')      
        plt.xlabel('X-axis')                 
        plt.ylabel('Y-axis')                 
        plt.legend()                         
        plt.grid(True) 
    
    def cdf(self, x):
        a = self.minimum
        b = self.maximum
        c = self.mode
        y_val = []
        if type(x) is list:
            for x_val in x:
                if a <= x_val and x_val <= c:
                    y_val.append ((x_val - a) ** 2 /((b - a) * (c - a)))
                elif c < x_val and x_val <= b:
                    y_val.append (1-((b - x_val) ** 2/ ((b - a) * (b - c))))
                else:
                     y_val.append(0.0)
            return y_val
        else:
            if a <= x and x <= c:
               return  (x - a) ** 2 /((b - a) * (c - a))
            elif c < x <= b:
                return (1- ((b - x) ** 2 / ((b - a) * (b - c))))
            else:
                return 0.0
    def plot_cdf(self):
        x = np.linspace(self.minimum, self.maximum, 10)
        x=x.tolist()
        y=TriangularDistribution.cdf(x)
        plt.plot(x, y, label='Graph', color='blue', linestyle='--') 
        plt.title('Graph')      
        plt.xlabel('X-axis')                 
        plt.ylabel('Y-axis')                 
        plt.legend()                         
        plt.grid(True) 
    
    def get_probability(self, x1,x2):
        a = self.minimum
        b = self.maximum
        c = self.mode
        if x1 <= a: x1_val =  0.0
        if x1 >= b: x1_val = 1.0
        if a <= x1 and x1 <= c:
            x1_val =  (x1 - a) ** 2 /((b - a) * (c - a))
        elif c < x1 and c <= b:
            x1_val = 1-(b - x1) ** 2 / ((b - a) * (b - c))
        if x2 <= a: x2_val =  0.0
        if x2 >= b: x2_val = 1.0
        if a <= x2 and x2 <= c:
            x2_val =  (x2 - a) ** 2 /((b - a) * (c - a))
        elif c < x2 and c <= b:
            x2_val = 1-(b - x2) ** 2 / ((b - a) * (b - c))
        return x2_val-x1_val
```

```python
def blend_images(image_A, image_B, weight_A, weight_B):
    print (f"Image A {image_A}")
    print (f"Image B {image_B}")
    image_A_weighted = image_A * weight_A
    image_B_weighted = image_B * weight_B
    print (f"Image A Weighted {image_A_weighted}")
    print (f"Image B Weighted {image_B_weighted}")
    image_c = image_A_weighted + image_B_weighted
    print (f"Image C by combining Image A and Image B after applying weighted average  {image_c}")



```

```python
image_A = np.random.randint(0, 255, size=(10, 10), dtype=np.uint8)
image_B = np.random.randint(0, 255, size=(10, 10), dtype=np.uint8)
weight_A = 2
weight_B = 3
blend_images(image_A, image_B, weight_A, weight_B)
```
