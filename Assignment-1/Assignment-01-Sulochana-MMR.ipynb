```python
def Second_Largest_Num (numbers):
    Second_Largest=sorted (numbers,reverse= True)
    return Second_Largest

```

### Question 2

```python
def Vowels_Count (Sentence):
    Sentence="Hello World"
    Count=0
    for x in Sentence :
        if x == "a" or x == "e" or x == "i" or x == "o" or x == "u" :
            Count=Count+1
    return Count
```

```python
Sentence="Hello World"
Count_Vowel = Vowels_Count (Sentence)
print(f"Vowel Count {Count_Vowel}")
```

```python
def rotate_list (Num_List,k):
    Num_List_FirstSlice = list(Num_List[-k:])
    i = k+1
    Num_List_SecondSlice = list(Num_List[:i])
    Num_List_Final = Num_List_FirstSlice + Num_List_SecondSlice
    print (F"{Num_List_Final}")  
```

### Question 8

```python
def group_by_category(products):
    from itertools import groupby
    products.sort(key=lambda x: x["category"])
    grouped_data = {}
    for key, group in groupby(products, key=lambda x: x["category"]):
        grouped_data[key] = list(group)
    print (F"{grouped_data}")







```

```python
products = [
    {"name": "Laptop", "price": 1200, "category": "Electronics"},
    {"name": "Book", "price": 25, "category": "Books"},
    {"name": "Tablet", "price": 300, "category": "Electronics"},
    {"name": "Fiction", "price": 15, "category": "Books"},
]
group_by_category(products)
```

## Question 10

```python
def reverse_the_words(sentence_text):
    sentence_list = sentence_text.split()
    sentence_list_len = len(sentence_list)
    Reverse_Sentence = ""
    for i in range (sentence_list_len-1,0,-1):
        if Reverse_Sentence == "":
            Reverse_Sentence = sentence_list[i] + " "
        else:
            Reverse_Sentence = Reverse_Sentence + sentence_list[i] + " " 
    print (F"{Reverse_Sentence}")
```

```python
reverse_the_words ("Python is awesome and challenging")
```
