### Question 9

```python
import pandas as pd
df = pd.read_csv("population_by_county.csv")
pop = pd.DataFrame(df)
filtered_column = pop.loc[pop['2010Census'] > 1000000, ['County']]
filtered_column_df = pd.DataFrame(filtered_column)
filtered_row_count = filtered_column_df.shape[0]

print (F"Number of Counties above 1 million in 2010 Census: {filtered_row_count}")
```

### Question 12

```python
import pandas as pd
df = pd.read_csv("population_by_county.csv")
pop = pd.DataFrame(df)
pop['Per_Diff']= round((((pop['2017PopEstimate'] - pop['2010Census'])/pop['2010Census'])*100),2)
filtered_column = pop.loc[pop['Per_Diff'] > 0, ['State','Per_Diff']]
filtered_column_df = pd.DataFrame(filtered_column).reset_index(drop=True).sort_values(by='Per_Diff', ascending= False)
print(F"{filtered_column_df}")
```
