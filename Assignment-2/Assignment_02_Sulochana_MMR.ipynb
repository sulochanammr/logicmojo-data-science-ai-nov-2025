### Question 5

```python
import numpy as np
num_rows = 8
num_cols = 8
k = 0
array_value =[0,1]
dynamic_array = np.zeros((num_rows, num_cols), dtype=int)
for i in range (0,num_rows):
    for j in range (0,num_cols):
        if k == 0:
            final_array = array_value [0]
            k=k+1
        else:
            final_array = array_value [1]
            k = 0

        dynamic_array[i, j] = final_array
print (F"{np.array(dynamic_array)}")
```
